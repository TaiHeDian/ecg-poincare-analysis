```python
import numpy as np
import matplotlib.pyplot as plt

from jvp_read import read_data
from ampd import ampd
from baseline_correct import baseline_correct
```

```python
commercial_datas = read_data('datas/jvp/commercial/')
gao_datas = read_data('datas/jvp/gao/')
lowsens_datas = read_data('datas/jvp/lowsens/')
```

```python
# Commercial Data 2.txt
df2 = commercial_datas['2.txt']
ax = df2.plot(x='Time/sec', y='Current/A', figsize=(10, 4))
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Current (A)')
plt.show()
```

```python
commercial_time2 = df2.to_numpy()[:,0]
commercial_data2 = df2.to_numpy()[:,1]
```

```python
peak_data2 = ampd(commercial_data2)
```
